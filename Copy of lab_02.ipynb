## Question 10
**Every Function:** Think of something you could store in a list. For example, you could make a list of mountains, rivers, countries, cities, languages, or anything else you’d like. Write a program that creates a list containing these items and then uses each function introduced in this chapter at least once.

```python
directions = ["North", "East", "West", "South"]
print("List:", directions)

#print each list item seperately
print("List items printed seperatly:")
print(directions[0])
print(directions[1])
print(directions[2])
print(directions[3])

#print each list item with text
print("List items printed with text:")
message1 = f"We are going the wrong way. We need to go " + (directions[0])
print(message1)
message2 = f"We are going the wrong way. We need to go " + (directions[1])
print(message2)
message3 = f"We are going the wrong way. We need to go " + (directions[2])
print(message3)
message4 = f"We are going the wrong way. We need to go " + (directions[3])
print(message4)

#insert item into list
directions.insert(0, "Northeast")
directions.append("Southwest")
print("List after added items:", directions)

#remove items with pop and del
directions.pop(4)
del directions[0]
print("List after deleted items:", directions)

#sorting list
print(sorted(directions))
print(directions)
print(sorted(directions, reverse = True))
print(directions)
directions.reverse()
print(directions)
directions.sort()
print(directions)
directions.sort(reverse=True)
print(directions)

#length
print(len(directions))
```
